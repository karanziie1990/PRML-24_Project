```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler


# Load CIFAR-10 dataset using TensorFlow
(train_images, train_labels), _ = tf.keras.datasets.cifar10.load_data()


# Flatten images based on RGB channels
num_samples = train_images.shape[0]  # Number of samples (images)
num_pixels = train_images.shape[1] * train_images.shape[2] * train_images.shape[3]  # Number of pixels per image

# Reshape the images to 1D arrays
flattened_images = train_images.reshape(num_samples, num_pixels)

# Print the shape of the flattened images
print("Shape of flattened images array:", flattened_images.shape)

# Flatten images based on RGB channels
num_samples = train_images.shape[0]  # Number of samples (images)
num_pixels = train_images.shape[1] * train_images.shape[2] * train_images.shape[3]  # Number of pixels per image

# Reshape the images to 1D arrays
flattened_images = train_images.reshape(num_samples, num_pixels)

# Create a DataFrame with pixel values and labels
df_images = pd.DataFrame(flattened_images)

# Add the label column to the DataFrame
df_images['Label'] = train_labels.flatten()

# Display the DataFrame
df_images
```

```python

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the features (excluding the 'Label' column)
scaler.fit(df_images.drop('Label', axis=1))

# Transform the features
scaled_features = scaler.transform(df_images.drop('Label', axis=1))

# Create a new DataFrame with the scaled features
df_images_scaled = pd.DataFrame(scaled_features, columns=df_images.columns[:-1])

# Add the 'Label' column to the scaled DataFrame
df_images_scaled['Label'] = df_images['Label']

# Display the DataFrame
df_images=df_images_scaled

df_images
```

```python
def dividingIntoClasses(df_images):
    """
    Creates separate DataFrames for each class based on the 'Label' column in the input DataFrame using Dictionary as a Dataform to store.

    Args:
        df_images (pandas.DataFrame): The input DataFrame containing image data.

    Returns:
        dict: A dictionary where keys are class labels and values are DataFrames containing data for each class.
    """
    class_dataframes = {}
    for class_label in range(10):
        class_dataframes[class_label] = df_images[df_images['Label'] == class_label]
    return class_dataframes

# Example usage:
# Assuming df_images is your original DataFrame
class_dataframes = dividingIntoClasses(df_images)
# Example: Access the DataFrame for class 'Airplane' (label 0)
airplane_df = class_dataframes[0]
print("Shape of Airplane DataFrame:", airplane_df.shape)
print(airplane_df)  # Display the first few rows
```
