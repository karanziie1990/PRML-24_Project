```python
def dividingIntoClasses(df_images):
    """
    Creates separate DataFrames for each class based on the 'Label' column in the input DataFrame using Dictionary as a Dataform to store.

    Args:
        df_images (pandas.DataFrame): The input DataFrame containing image data.

    Returns:
        dict: A dictionary where keys are class labels and values are DataFrames containing data for each class.
    """
    class_dataframes = {}
    for class_label in range(10):
        class_dataframes[class_label] = df_images[df_images['Label'] == class_label]
    return class_dataframes

# Example usage:
# Assuming df_images is your original DataFrame
class_dataframes = dividingIntoClasses(df_images)
# Example: Access the DataFrame for class 'Airplane' (label 0)
airplane_df = class_dataframes[0]
print("Shape of Airplane DataFrame:", airplane_df.shape)
airplane_df  # Display the first few rows
```

```python
def calculate_means(class_dataframes):
    """
    Calculates the mean of each column for each class DataFrame.

    Args:
        class_dataframes (dict): A dictionary where keys are class labels and values are DataFrames containing data for each class.

    Returns:
        dict: A dictionary where keys are class labels and values are Series containing the mean of each column.
    """
    class_means = {}
    for class_label, df in class_dataframes.items():
        # Calculate the mean of each column (excluding the 'Label' column)
        class_means[class_label] = df.drop('Label', axis=1).mean()
    return class_means

# Example usage:
# Assuming class_dataframes is your dictionary of class DataFrames
class_means = calculate_means(class_dataframes)
# Example: Access the mean values for class 'Airplane' (label 0)
airplane_means = class_means[0]
print("Mean values for Airplane class:")
airplane_means
```

```python
def S_w(class_dataframes):
    """
    Calculates the within-class scatter matrix and displays each scaled covariance matrix.

    Args:
        class_dataframes (dict): A dictionary where keys are class labels and values are DataFrames containing data for each class.

    Returns:
        pandas.DataFrame: The within-class scatter matrix.
    """
    # Initialize an empty DataFrame to store the within-class scatter matrix
    scatter_matrix = pd.DataFrame(np.zeros((class_dataframes[0].shape[1]-1, class_dataframes[0].shape[1]-1)))

    for class_label, df in class_dataframes.items():
        # Drop the 'Label' column
        df_features = df.drop('Label', axis=1)

        # Calculate the covariance matrix of the features
        cov_matrix = df_features.cov()

        # Multiply the covariance matrix by the number of samples in the class
        scaled_cov_matrix = cov_matrix * df_features.shape[0]

        # # Display the scaled covariance matrix
        # print(f"Scaled Covariance Matrix for Class {class_label}:")
        # print(scaled_cov_matrix)
        # print("\n" + "="*50 + "\n")  # Print a separator

        # Add the scaled covariance matrix to the scatter matrix
        scatter_matrix += scaled_cov_matrix

    return scatter_matrix

# Example usage:
# Assuming class_dataframes is your dictionary of class DataFrames
within_class_scatter_matrix = S_w(class_dataframes)
print("Within-Class Scatter Matrix:")
within_class_scatter_matrix
```

```python
def S_b(class_dataframes, class_means):
    """
    Calculates the between-class scatter matrix.

    Args:
        class_dataframes (dict): A dictionary where keys are class labels and values are DataFrames containing data for each class.
        class_means (dict): A dictionary where keys are class labels and values are Series containing the mean of each column.

    Returns:
        pandas.DataFrame: The between-class scatter matrix.
    """
    # Calculate the overall mean of all classes
    overall_mean = pd.concat(class_dataframes.values()).drop('Label', axis=1).mean()

    # Initialize an empty DataFrame to store the between-class scatter matrix
    scatter_matrix = pd.DataFrame(np.zeros((class_dataframes[0].shape[1]-1, class_dataframes[0].shape[1]-1)))

    for class_label, df in class_dataframes.items():
        # Calculate the mean difference (mean vector - overall mean)
        mean_diff = class_means[class_label] - overall_mean

        # Multiply the outer product of the mean difference by the number of samples in the class
        scaled_outer_product = np.outer(mean_diff, mean_diff) * df.shape[0]

        # Add the scaled outer product to the scatter matrix
        scatter_matrix += scaled_outer_product

    return scatter_matrix

# Example usage:
# Assuming class_dataframes is your dictionary of class DataFrames
# and class_means is your dictionary of class means
between_class_scatter_matrix = S_b(class_dataframes, class_means)
print("Between-Class Scatter Matrix:")
between_class_scatter_matrix
```

```python

# Calculate the inverse of the within-class scatter matrix
Sw_inv = np.linalg.inv(within_class_scatter_matrix)

# Multiply the inverse of the within-class scatter matrix with the between-class scatter matrix
result = np.dot(Sw_inv, between_class_scatter_matrix)

# Convert the result to a DataFrame
result_df = pd.DataFrame(result)

# Display the DataFrame
result_df
```

```python
import numpy as np

def sorted_eigenvectors(df):
    """
    Calculates the eigenvalues and eigenvectors of a DataFrame and returns the eigenvectors sorted by their corresponding eigenvalues in descending order.

    Args:
        df (pandas.DataFrame): The input DataFrame.

    Returns:
        list: A list of tuples where each tuple contains an eigenvalue and its corresponding eigenvector. The list is sorted in descending order of the eigenvalues.
    """
    # Calculate the eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(df)

    # Create a list of tuples where each tuple is an eigenvalue and its corresponding eigenvector
    eig_pairs = [(np.abs(eigenvalues[i]), eigenvectors[:, i]) for i in range(len(eigenvalues))]

    # Sort the list of tuples in descending order of the eigenvalues
    eig_pairs.sort(key=lambda x: x[0], reverse=True)

    return eig_pairs

# Example usage:
# Assuming result_df is your DataFrame
eig_pairs = sorted_eigenvectors(result_df)
# Example: Access the eigenvector corresponding to the largest eigenvalue
largest_eigenvector = eig_pairs[1][1]
print("Eigenvector corresponding to the largest eigenvalue:")
print(largest_eigenvector)

print("Shape of the largest eigenvector:", largest_eigenvector.shape)
```

```python
def reduce_and_add_data_labels(class_dataframes, eig_pairs, k):
    """
    Reduces the dimensionality of each DataFrame in class_dataframes from n to k using the top k eigenvectors
    and adds the last column from each original DataFrame to the reduced DataFrame.

    Args:
        class_dataframes (dict): A dictionary where keys are class labels and values are original DataFrames containing data for each class.
        eig_pairs (list): A list of tuples where each tuple contains an eigenvalue and its corresponding eigenvector. The list is sorted in descending order of the eigenvalues.
        k (int): The number of dimensions to reduce to.

    Returns:
        dict: A dictionary where keys are class labels and values are DataFrames containing data for each class, with reduced dimensionality and an additional column.
    """
    # Get the top k eigenvectors
    top_k_eigenvectors = np.array([eig_pairs[i][1] for i in range(k)]).T

    # Initialize an empty dictionary to store the reduced DataFrames
    reduced_dataframes = {}

    for class_label, df in class_dataframes.items():
        # Drop the 'Label' column
        df_features = df.drop('Label', axis=1)

        # Perform the dimensionality reduction
        reduced_features = np.dot(df_features, top_k_eigenvectors)

        # Create a new DataFrame with the reduced features
        reduced_df = pd.DataFrame(reduced_features)

        # Store the last column from the original DataFrame
        data_labels = df.iloc[:, -1]

        # Add the last column to the reduced DataFrame
        reduced_df['data_labels'] = data_labels.values

        # Store the reduced DataFrame in the dictionary
        reduced_dataframes[class_label] = reduced_df

    return reduced_dataframes

# Example usage:
# Assuming class_dataframes is your dictionary of class DataFrames
# and eig_pairs is your list of eigenvalue-eigenvector pairs
k = 50  # The number of dimensions to reduce to
reduced_dataframes = reduce_and_add_data_labels(class_dataframes, eig_pairs, k)
# Example: Access the reduced DataFrame for class 'Airplane' (label 0)
reduced_airplane_df = reduced_dataframes[0]
print("Shape of reduced Airplane DataFrame:", reduced_airplane_df.shape)
reduced_airplane_df  # Display the first few rows
```
